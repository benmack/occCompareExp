## Setup experiments

A proper set up of the experiments is required by defining the classifier comparison settings list ("parcc2). Furthermore, a function (``get_refset()``) must be available which creates the reference sets as required by the functions of occCompaRe

In doubt how this can be done see "_nb_02_setup_experiments.ipynb".

```r
parcc <- set_expOpts("revision1", "b")
```

## Run the experimentsâ€¦ (takes a lot of time!)

The ``get_refset()`` as defined here returns a training, validation and test data set. The following function call creates a model, predicts the validation and test data and evaluates the models based on the best achievable threshold (``fun=.trainUtest``). 
Then it calculates the PU-performance metrics based on the validation samples (``fun=.resampling-val``). 
The function itself returns ``NULL`` but all the results are written to the directory given in ``parcc$dn.res``.

```r
run_experiments(parcc=parcc, 
                fun=c(".trainUtest", 
                      ".resampling-val"), doPar=FALSE)
```

Finally, ``add_thVals2resTables`` adds results at some additional thresholds to the results tables. 
This function loads the files found by ``get_summaryFile()`` and adds the accuracies for the thresholds "minTP_T", "tenTP_T", "mxSSSP_T", "eqSSSP_T" which are derived from the PU validation data.

Note that the function modifies the files in:

```r
file.path(parcc$dn.res, "resTables")
```

But befor it creates a copy of the file in the directory:

```r
file.path(parcc$dn.res, "resTables", "resTables_BEFORE_add_thVals2resTable_<date-time string>")
```

If you want to extend the parameter grids it is necessary to replace the files in the "resTables" folder with the bckuped ones (because else the results of the new models cannot be written to the summary files), , run ``run_experiments()`` and then again ``add_thVals2resTable()``.

```r
add_thVals2resTable()
```
